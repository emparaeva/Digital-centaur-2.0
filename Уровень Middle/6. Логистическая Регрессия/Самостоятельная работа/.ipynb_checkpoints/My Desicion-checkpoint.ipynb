Информация о клиенте
- 1 - age (numeric) - возраст
- 2 - job - тип работы
- 3 - marital - семейное положение
- 4 - education - уровень образования
- 5 – balance: Уровень дохода
- 6 - housing: Есть ли ипотека
- 7 - loan: Есть ли кредит

Информация по последнему контакту в текущей кампании: 
- 8 - contact: Тип коммуникации
- 9 - day_of_week: день последнего контакта с клиентом
- 10 - month: месяц последнего контакта с клиентом
- 11 - duration: длительность последнего контакта, в секундах (numeric). 

Прочие признаки: 
- 12 - campaign: количество контактов с клиентом (numeric, includes last contact)
- 13 - pdays: количество дней, которое прошло между последним и предпоследним контактом
- 14 - previous: количество контактов с клиентом, исключая последний контакт (numeric)
- 15 - poutcome: результаты последней рекламной компании 

Целевая переменная:
- 16 - y - откроет ли клиент инвестиционный счет в нашем банке?  (binary: 'yes','no')

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
label_encoder = LabelEncoder()
one_hot_enc = OneHotEncoder()
```

    Применим Label Encoder к тестовой и обучающей выборке:

```python
for feature in cat_features_enc:
    df_train_points[feature] = label_encoder.fit_transform(df_train_points[feature])
    df_test[feature] = label_encoder.transform(df_test[feature])
```

    Применим One Hot Encoding к cat_features_hot

```python
df_train_points_onehot = one_hot_enc.fit_transform(df_train_points[cat_features_hot]).toarray()
df_test_onehot = one_hot_enc.transform(df_test[cat_features_hot]).toarray()
```

```python
columns = one_hot_enc.get_feature_names(cat_features_hot)
columns
```

```python
df_train_points_onehot = pd.DataFrame(df_train_points_onehot, columns=columns)
df_test_onehot = pd.DataFrame(df_test_onehot, columns=columns)
```

```python
# соединяем новую табличку с исходной
df_train_points_new = pd.concat([df_train_points, df_train_points_onehot], axis=1)
df_test_new = pd.concat([df_test, df_test_onehot], axis=1)

# удаляем старые категориальные признаки
df_train_points_new = df_train_points_new.drop(columns=cat_features_hot)
df_test_new = df_test_new.drop(columns=cat_features_hot)
```

    Валидация. Кросс-валидация. Обучение модели

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
```

```python
# метрики для задачи классификации
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
```

    Обычная валидация

```python
X_train, X_test, y_train, y_test = train_test_split(df_train_points_new, y, test_size=0.3)
```

    KNN и кросс-валидация

```python
knn = KNeighborsClassifier(n_neighbors=105)

knn.fit(X_train, y_train)
knn_preds = knn.predict(X_test)
knn_preds_proba = knn.predict_proba(X_test)[:, 1]

print('accuracy', accuracy_score(y_test, knn_preds))
print('roc_auc', roc_auc_score(y_test, knn_preds_proba))

cross_val_scores = cross_val_score(knn, df_train_points_new, y, cv=10, scoring='roc_auc')
print(cross_val_scores)
print(np.mean(cross_val_scores))
```

    Logistic Regression

```python
lr = LogisticRegression()

lr.fit(X_train, y_train)
lr_preds = lr.predict(X_test)
lr_preds_proba = lr.predict_proba(X_test)[:, 1]

print('accuracy', accuracy_score(y_test, lr_preds))
print('roc_auc', roc_auc_score(y_test, lr_preds_proba))

cross_val_scores = cross_val_score(lr, df_train_points_new, y, cv=10, scoring='roc_auc')
print(cross_val_scores)
print(np.mean(cross_val_scores))
```

    GBDT

```python
gbdt = GradientBoostingClassifier() # инициализируем gbdt

gbdt.fit(X_train, y_train)
gbdt_preds = gbdt.predict(X_test)
gbdt_preds_proba = gbdt.predict_proba(X_test)[:, 1]

print('accuracy', accuracy_score(y_test, gbdt_preds))
print('roc_auc', roc_auc_score(y_test, gbdt_preds_proba))

cross_val_scores = cross_val_score(gbdt, df_train_points_new, y, cv=10, scoring='roc_auc')
print(cross_val_scores)
print(np.mean(cross_val_scores))
```

    GridSearchCV

```python
param_grid = {
    "max_depth": [3, 4, 5], # глубина дерева
    "n_estimators": [50, 100, 150], # кол-во деревьев
    "learning_rate": [0.01, 0.05, 150],
    "subsample": [0.8, 0.9, 1] 
}
```

```python
gbdt = GradientBoostingClassifier()
kfold = KFold(n_splits=6, shuffle=True, random_state=123)
```

```python
CV_gbdt = GridSearchCV(estimator=gbdt,
                       param_grid=param_grid,
                       scoring='roc_auc',
                       cv=kfold,
                       verbose = 1000)
```

```python
CV_gbdt.fit(df_train_points_new, y)
```

```python
CV_gbdt.best_params_
```

```python
CV_gbdt.best_score_
```

```python
y_pred_gbdt = CV_gbdt.predict(df_test_new)
```
